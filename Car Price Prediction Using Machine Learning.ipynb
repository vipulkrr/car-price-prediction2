# now creating Feature Matrix to our dataset for regression to use machine learning algorithm

# Store Feature Matrix in X and Response ( Target ) in Vector y

```python
# selecting feature
x = data.drop(['name','selling_price'],axis=1)
y = data['selling_price']
```

```python
y
```

```python
x
```

# Splitting the Dataset into the Training Set and Test Set

```python
# importing sklearn library to run training and testing algorithm
from sklearn.model_selection import train_test_split
```

```python
# Split the dataset into training and test set
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
data.head()
```

# Import the Models

```python
import xgboost as xgb
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from xgboost import XGBRegressor
```

# Model Training

```python
lr = LinearRegression()
lr.fit(x_train,y_train)

rf = RandomForestRegressor()
rf.fit(x_train,y_train)

xgb = GradientBoostingRegressor()
xgb.fit(x_train,y_train)

xg = XGBRegressor()
xg.fit(x_train,y_train)
```

#  Prediction on Test Data

```python
y_pred1 = lr.predict(x_test)
y_pred2 = rf.predict(x_test)
y_pred3 = xgb.predict(x_test)
y_pred4 = xg.predict(x_test)
```

# Evaluating the Algorithm

```python
from sklearn import metrics
```

```python
score1 = metrics.r2_score(y_test,y_pred1)
score2 = metrics.r2_score(y_test,y_pred2)
score3 = metrics.r2_score(y_test,y_pred3)
score4 = metrics.r2_score(y_test,y_pred4)
```

```python
print(score1,score2,score3,score4)
```

```python
# finally execute our all models and just checked it out which is good
final_data = pd.DataFrame({'Models':['LR','RF','GBR','XG'],"R2_SCORE":[score1,score2,score3,score4]})
```

```python
final_data
```

```python
colors = ['red','green','blue','yellow']
# lets visualized this
sns.barplot(x=final_data['Models'],y=final_data['R2_SCORE'],data=final_data,palette=colors)
```
